### 2. Create groupby with number of trips

```python
df_group = df.groupby(['start_station_name', 'end_station_name'])['value'].count().reset_index()
```

```python
df_group
```

```python
print(df_group['value'].sum())
print(df.shape)
```

```python
df_group.rename(columns = {'value': 'trips'}, inplace = True)
```

```python
df_group
```

### 3. Create start stations dataset for merge

```python
ss = df_group["start_station_name"].unique()
ss.shape
```

```python
start_name = df_group[["start_station_name"]]
start_name.drop_duplicates(inplace = True)
print(start_name.shape)
start_stations = df[['start_station_name', 'start_lat','start_lng']]
start_stations.drop_duplicates(inplace = True)
print(start_stations.shape)
```

```python
list_start_stations = []
```

```python
for i in range(len(start_name)): # 1761
  station_name = start_name.iloc[i]['start_station_name'] # start_name.iloc[0]['start_station_name']
  if station_name in start_stations['start_station_name'].values:
    data = start_stations[start_stations['start_station_name'] == station_name]
    list_start_stations.append(data)
```

```python
start_station_data = pd.concat(list_start_stations).reset_index(drop = True)
```

```python
start_station_data.head()
```

```python
start_station_data.drop_duplicates(subset=['start_station_name'],inplace = True)
```

```python
start_station_data.reset_index(drop = True, inplace = True)
```

```python
start_station_data.head()
```

```python
df_partial_data = df_group.merge(start_station_data, left_on = 'start_station_name', right_on = 'start_station_name', how='left')
```

```python
df_partial_data.head()
```

### 4. Create end stations dataset for merge

```python
end_name = df_group[["end_station_name"]]
end_name.drop_duplicates(inplace = True)
print(end_name.shape)
end_stations = df[['end_station_name', 'end_lat','end_lng']]
end_stations.drop_duplicates(inplace = True)
print(end_stations.shape)
```

```python
list_end_stations = []
for i in range(len(end_name)): # 1761
  station_name = end_name.iloc[i]['end_station_name'] # end_name.iloc[0]['end_station_name']
  if station_name in end_stations['end_station_name'].values:
    data = end_stations[end_stations['end_station_name'] == station_name]
    list_end_stations.append(data)
```

```python
end_station_data = pd.concat(list_end_stations).reset_index(drop = True)
end_station_data.head()
end_station_data.drop_duplicates(subset=['end_station_name'],inplace = True)
end_station_data.reset_index(drop = True, inplace = True)
end_station_data.head()
```

### 5. Merge final dataset for keplerGL

```python
df_station_data = df_partial_data.merge(end_station_data, left_on = 'end_station_name', right_on = 'end_station_name', how='left')
```

```python
df_station_data.head()
```

```python
df_station_data.to_pickle('/content/drive/MyDrive/Colab Notebooks/df_station_data.pickle')
```

### 6. Create and save KeplerGL map

```python
m = KeplerGl(height=700, data={'data_1': df_station_data})
m
```

```python
config = m.config
```

```python
import json
with open('/content/drive/MyDrive/Colab Notebooks/config.json', 'w') as f:
  json.dump(config, f)
```

```python
m.save_to_html(file_name='NewYork City Bikes Trip Map.html', read_only=False, config=config)
```
